````python
rand_data=np.random.random(20)
rand_data[3]*=3
rand_data[7]*=3
rand_data[10]*=3
rand_data[14]*=-3
rand_data[17]*=-3
rand_data[19]*=-3
rand_data
plt.figure(figsize=(4,6))
bp=plt.boxplot(rand_data, tick_labels=['Data'], vert=True, patch_artist=True);
# Labeling Outliers
outliers = bp['fliers'][0].get_ydata()
for outlier in outliers:
    # plt.text(1.1, outlier, f'{outlier:.2f}', verticalalignment='center')
    plt.text(1.1, outlier, f'outliers', verticalalignment='center')

# Labeling Whiskers
whiskers = bp['whiskers']
for i, whisker in enumerate(whiskers):
    y_data = whisker.get_ydata()
    if i % 2 == 0:  # Lower whisker
        # plt.text(1.1, y_data[1], f'Lower Whisker: {y_data[1]:.2f}', verticalalignment='center')
        plt.text(1.1, y_data[1], f'Q1 - 1.5 IQR', verticalalignment='center')
    else:  # Upper whisker
        # plt.text(1.1, y_data[1], f'Upper Whisker: {y_data[1]:2f}', verticalalignment='center')
        plt.text(1.1, y_data[1], f'Q1 + 1.5 IQR', verticalalignment='center')


# Labeling Box Parts
medians = bp['medians']
for median in medians:
    y_data = median.get_ydata()
    # plt.text(1.1, y_data[0], f'Median: {y_data[0]:.2f}', verticalalignment='center')
    plt.text(1.1, y_data[0], f'Q2', verticalalignment='center')

# Labeling Box Parts (Corrected)
boxes = bp['boxes']
for box in boxes:
    path = box.get_path()
    vertices = path.vertices
    y_vertices = vertices[:, 1]  # Extract y-coordinates
    # plt.text(1.1, y_vertices.min(), f'Q1: {y_vertices.min():.2f}', verticalalignment='center')
    # plt.text(1.1, y_vertices.max(), f'Q3: {y_vertices.max():.2f}', verticalalignment='center')
    plt.text(1.1, y_vertices.min(), f'Q1', verticalalignment='center')
    plt.text(1.1, y_vertices.max(), f'Q3', verticalalignment='center')
````

````python
# Sample data
categories = ['A', 'B', 'C', 'D']
data1 = [20, 35, 30, 45]
data2 = [15, 25, 40, 20]
data3 = [10, 18, 22, 30]

# Create the stacked bar chart
plt.figure(figsize=(8, 6))

# Create the bars and store them for labeling
bars1 = plt.bar(categories, data1, label='Data 1')
bars2 = plt.bar(categories, data2, bottom=data1, label='Data 2')
bars3 = plt.bar(categories, data3, bottom=[sum(x) for x in zip(data1, data2)], label='Data 3')

# Add labels to each segment of the stacked bars
for i, category in enumerate(categories):
    plt.text(category, data1[i] / 2, str(data1[i]), ha='center', va='center')
    plt.text(category, data1[i] + data2[i] / 2, str(data2[i]), ha='center', va='center')
    plt.text(category, data1[i] + data2[i] + data3[i] / 2, str(data3[i]), ha='center', va='center')

# Add labels to the axes and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart with Labels')

# Add a legend
plt.legend()

# Show the plot
plt.show()
````
